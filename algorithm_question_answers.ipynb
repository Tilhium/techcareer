```python
#find the biggestlist element

def find_biggest(list):
    max_number=list[0]
    for number in list:
        if number > max_number:
            max_number=number
    return max_number


print(find_biggest([1,2,32,4,5,6,7]))

            

    
```

```python
#factorial calculating

def factorial():
    number = int(input("Pls enter your number: "))
    if number == 0:
        return 1
    result = 1

    for i in range (1,number+1):
        result*=i
    return result

print("factorial result: ",factorial())
```

```python
def union_unique(list_a, list_b):
    return list(dict.fromkeys(list(list_a) + list(list_b)))
```

```python
def is_palindrome(text):
    s = "".join(ch.lower() for ch in text if ch.isalnum())
    return s == s[::-1]
```

```python
def gcd_ebob(x, y):
    a, b = x, y
    while b:
        a, b = b, a % b
    return a
```

```python
def is_anagram(s1, s2):
    a = "".join(ch.lower() for ch in s1 if ch.isalnum())
    b = "".join(ch.lower() for ch in s2 if ch.isalnum())
    return sorted(a) == sorted(b)
```

```python
def find_missing_number(nums):
    n = len(nums) + 1
    total = n * (n + 1) // 2
    return total - sum(nums)
```

```python
def rotate_list(arr, k):
    if not arr:
        return arr
    k = k % len(arr)
    return arr[-k:] + arr[:-k]
```

```python
def median_two_sorted(nums1, nums2):
    A, B = nums1, nums2
    m, n = len(A), len(B)
    if m > n:
        A, B, m, n = B, A, n, m
    if n == 0:
        raise ValueError("Both arrays empty")
    imin, imax, half = 0, m, (m + n + 1) // 2
    while imin <= imax:
        i = (imin + imax) // 2
        j = half - i
        Aleft = A[i-1] if i > 0 else float("-inf")
        Aright = A[i] if i < m else float("inf")
        Bleft = B[j-1] if j > 0 else float("-inf")
        Bright = B[j] if j < n else float("inf")
        if Aleft <= Bright and Bleft <= Aright:
            if (m + n) % 2 == 1:
                return float(max(Aleft, Bleft))
            return (max(Aleft, Bleft) + min(Aright, Bright)) / 2.0
        elif Aleft > Bright:
            imax = i - 1
        else:
            imin = i + 1
    raise ValueError("Input arrays not sorted or invalid")
```

```python
def longest_palindrome_substring(s):
    if not s:
        return ""
    start, end = 0, 0
    for i in range(len(s)):
        l, r = i, i
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if r - l > end - start:
                start, end = l, r
            l -= 1; r += 1
        l, r = i, i + 1
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if r - l > end - start:
                start, end = l, r
            l -= 1; r += 1
    return s[start:end+1]
```

```python
from collections import deque, defaultdict

def word_ladder_length(begin_word, end_word, word_list):
    word_set = set(word_list)
    if end_word not in word_set:
        return 0
    L = len(begin_word)
    combo = defaultdict(list)
    for word in word_set:
        for i in range(L):
            combo[word[:i] + "*" + word[i+1:]].append(word)
    queue = deque([(begin_word, 1)])
    visited = {begin_word}
    while queue:
        word, level = queue.popleft()
        for i in range(L):
            pattern = word[:i] + "*" + word[i+1:]
            for adj in combo.get(pattern, []):
                if adj == end_word:
                    return level + 1
                if adj not in visited:
                    visited.add(adj)
                    queue.append((adj, level + 1))
            combo[pattern] = []
    return 0
```

```python
def max_subarray_sum(nums):
    best = current = nums[0]
    for x in nums[1:]:
        current = x if x > current + x else current + x
        best = current if current > best else best
    return best
```

```python
def merge_intervals(intervals):
    if not intervals:
        return []
    intervals_sorted = sorted(intervals, key=lambda x: x[0])
    merged = [intervals_sorted[0]]
    for start, end in intervals_sorted[1:]:
        last_start, last_end = merged[-1]
        if start <= last_end:
            merged[-1] = [last_start, max(last_end, end)]
        else:
            merged.append([start, end])
    return merged






```
